```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
red_wine = pd.read_csv('WineQuality-RedWine.csv')
white_wine = pd.read_csv('WineQuality-WhiteWine.csv')
```

```python
# Check for missing values in both datasets
missing_values_red = red_wine.isnull().sum()
missing_values_white = white_wine.isnull().sum()

missing_values_red
```

```python
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler

# Initialize the scaler
scaler = StandardScaler()

# Scale the datasets
red_wine_scaled = scaler.fit_transform(red_wine.drop('quality', axis=1))
white_wine_scaled = scaler.fit_transform(white_wine.drop('quality', axis=1))

# Extract target variables
red_wine_target = red_wine['quality']
white_wine_target = white_wine['quality']

# Split the datasets into training and testing sets
red_train_features, red_test_features, red_train_target, red_test_target = train_test_split(
    red_wine_scaled, red_wine_target, test_size=0.2, random_state=42)

white_train_features, white_test_features, white_train_target, white_test_target = train_test_split(
    white_wine_scaled, white_wine_target, test_size=0.2, random_state=42)

# Output the shapes of the splits to verify
(red_train_features.shape, red_test_features.shape, white_train_features.shape, white_test_features.shape)
```

```python
from sklearn.linear_model import LinearRegression
from sklearn.ensemble import RandomForestRegressor, GradientBoostingRegressor
from sklearn.metrics import mean_squared_error, r2_score

# Initialize the models
linear_model = LinearRegression()
random_forest_model = RandomForestRegressor(random_state=42)
gbm_model = GradientBoostingRegressor(random_state=42)

# Train the Linear Regression model
linear_model.fit(red_train_features, red_train_target)
linear_predictions = linear_model.predict(red_test_features)
linear_mse = mean_squared_error(red_test_target, linear_predictions)
linear_r2 = r2_score(red_test_target, linear_predictions)

# Train the Random Forest model
random_forest_model.fit(red_train_features, red_train_target)
rf_predictions = random_forest_model.predict(red_test_features)
rf_mse = mean_squared_error(red_test_target, rf_predictions)
rf_r2 = r2_score(red_test_target, rf_predictions)

# Train the GBM model
gbm_model.fit(red_train_features, red_train_target)
gbm_predictions = gbm_model.predict(red_test_features)
gbm_mse = mean_squared_error(red_test_target, gbm_predictions)
gbm_r2 = r2_score(red_test_target, gbm_predictions)

(linear_mse, linear_r2, rf_mse, rf_r2, gbm_mse, gbm_r2)
```

```python
from sklearn.model_selection import GridSearchCV

# Define the parameter grid for Random Forest
rf_param_grid = {
    'n_estimators': [100, 200, 300],
    'max_features': ['auto', 'sqrt'],
    'max_depth': [10, 20, None],
    'min_samples_split': [2, 5, 10]
}

# Setup the grid search
rf_grid_search = GridSearchCV(estimator=random_forest_model, param_grid=rf_param_grid, cv=3, scoring='neg_mean_squared_error', verbose=1)

# Perform grid search
rf_grid_search.fit(red_train_features, red_train_target)
```

```python
# Revised and simplified parameter grid for Random Forest
rf_param_grid_revised = {
    'n_estimators': [100, 150],  # Fewer options
    'max_depth': [None, 20],     # Fewer options
    'min_samples_split': [2, 5]  # Fewer options
}

# Setup the grid search with reduced folds for faster computation
rf_grid_search_revised = GridSearchCV(estimator=random_forest_model, param_grid=rf_param_grid_revised, cv=2, scoring='neg_mean_squared_error', verbose=1)

# Perform the revised grid search
rf_grid_search_revised.fit(red_train_features, red_train_target)
```

```python
# Get the best parameters and best score (neg_mean_squared_error)
best_params = rf_grid_search_revised.best_params_
best_score = rf_grid_search_revised.best_score_

# Evaluate the best model on the test data
best_rf_model = rf_grid_search_revised.best_estimator_
best_rf_predictions = best_rf_model.predict(red_test_features)
best_rf_mse = mean_squared_error(red_test_target, best_rf_predictions)
best_rf_r2 = r2_score(red_test_target, best_rf_predictions)

(best_params, best_score, best_rf_mse, best_rf_r2)
```

```python
# Train the Linear Regression model on white wine data
linear_model.fit(white_train_features, white_train_target)
linear_predictions_white = linear_model.predict(white_test_features)
linear_mse_white = mean_squared_error(white_test_target, linear_predictions_white)
linear_r2_white = r2_score(white_test_target, linear_predictions_white)

# Train the Random Forest model on white wine data
random_forest_model.fit(white_train_features, white_train_target)
rf_predictions_white = random_forest_model.predict(white_test_features)
rf_mse_white = mean_squared_error(white_test_target, rf_predictions_white)
rf_r2_white = r2_score(white_test_target, rf_predictions_white)

# Train the GBM model on white wine data
gbm_model.fit(white_train_features, white_train_target)
gbm_predictions_white = gbm_model.predict(white_test_features)
gbm_mse_white = mean_squared_error(white_test_target, gbm_predictions_white)
gbm_r2_white = r2_score(white_test_target, gbm_predictions_white)

(linear_mse_white, linear_r2_white, rf_mse_white, rf_r2_white, gbm_mse_white, gbm_r2_white)
```

## Red Wine

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Feature distribution for red wine data
plt.figure(figsize=(12, 10))
for i, col in enumerate(red_wine.columns[:-1], 1):  # Exclude 'quality' from distributions
    plt.subplot(4, 3, i)
    sns.histplot(red_wine[col], kde=True, color='red')
    plt.title(f'Distribution of {col}')
plt.tight_layout()
plt.show()

# Correlation matrix heatmap
plt.figure(figsize=(10, 8))
correlation_matrix = red_wine.corr()
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', fmt=".2f")
plt.title('Correlation Matrix')
plt.show()
```

```python
# Corrected scatter plot for Actual vs Predicted values by Random Forest
plt.figure(figsize=(10, 6))
sns.scatterplot(x=red_test_target, y=rf_predictions, alpha=0.6, color='red')
plt.plot([red_test_target.min(), red_test_target.max()], [red_test_target.min(), red_test_target.max()], 'k--', lw=3)
plt.xlabel('Actual Quality')
plt.ylabel('Predicted Quality')
plt.title('Actual vs Predicted Quality (Random Forest)')
plt.show()

# Residual plot for the Random Forest model
residuals = red_test_target - rf_predictions
plt.figure(figsize=(10, 6))
sns.histplot(residuals, kde=True, color='red')
plt.title('Distribution of Residuals (Random Forest)')
plt.xlabel('Residuals')
plt.ylabel('Frequency')
plt.show()
```

```python
from sklearn.metrics import confusion_matrix, accuracy_score
import numpy as np

# Helper function to categorize quality
def categorize_quality(quality):
    if quality <= 4:
        return 'Low'
    elif quality <= 6:
        return 'Medium'
    else:
        return 'High'

# Apply categorization to actual and predicted quality scores
actual_categories = red_test_target.apply(categorize_quality)
predicted_categories = pd.Series(rf_predictions).apply(np.round).apply(categorize_quality)  # Round predictions to nearest int

# Compute the confusion matrix
conf_matrix = confusion_matrix(actual_categories, predicted_categories, labels=['Low', 'Medium', 'High'])

# Compute accuracy
accuracy = accuracy_score(actual_categories, predicted_categories)

conf_matrix, accuracy
```

```python
# Visualization of the confusion matrix
plt.figure(figsize=(8, 6))
sns.heatmap(conf_matrix, annot=True, fmt='d', cmap='Blues', xticklabels=['Low', 'Medium', 'High'], yticklabels=['Low', 'Medium', 'High'])
plt.title('Confusion Matrix for Red Wine')
plt.xlabel('Predicted Class')
plt.ylabel('Actual Class')
plt.show()
```
